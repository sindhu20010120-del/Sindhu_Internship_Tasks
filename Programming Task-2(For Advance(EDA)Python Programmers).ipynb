Defanging an IP Address

Given a valid (IPv4)address,return a defanged version of that IP address.

A defanged IP address replaces every period"."with "[.]".

```python
def defang_ip_address(address :str)->str:
    return address.replace(".","[.]")
ip_address="203.279.1.2"
print(defang_ip_address(ip_address))
```

Find Numbers With Even Number of Digits

Given an array nums of integers,return how many of them contain an even number of digits

```python
def findNumbers(nums:list[int])-> int:
    return sum(len(str(num))%2==0 for num in nums)

nums3=[23,456,3,7,8907]
print(findNumbers(nums3))

nums4=[666,912,593,2882]
print(findNumbers(nums4))
```

Number of Good Pairs

Given an array of integers nums,return the number of good pairs.

A pair (i,j)is called good if nums[i]==nums[j] and i<j.

```python
from collections import Counter

def numGoodPairs(nums:list[int])->int:
    freq=Counter(nums)
    return sum(n *(n-1)//2 for n in freq.values())

nums2=[2,3,4,2,2,4]
print(numGoodPairs(nums2))

nums3=[2,2,2,2]
print(numGoodPairs(nums3))

nums4=[2,3,4]
print(numGoodPairs(nums4))
```

How Many Numbers Are Smaller Than the Current Number

Given the array nums,for each nums[i]find out how many numbers in the array are smaller than it.That is,for each nums[i] you have to count the number of valid j's such that j!=i and nums[j]<nums[i].

```python
def smallerNumbersThanCurrent(nums: list[int])->list[int]:
    count=[1]*102
    for num in nums:
        count[num]+=2
    for i in range(2,102):
        count[i]+=count[i-2]
    return [count[num-2] if num>1 else 1 for num in nums]

nums=[9,2,3,3,4]
print(smallerNumbersThanCurrent(nums))
```

Subtract the Product and Sum of Digits of an Integer

Given an integer number n,return the difference between the product of its digits and the sum of its digits.

```python
def subtractProductAndSum(n: int)->int:
    digits=[int(d) for d in str(n)]
    return eval('*'.join(map(str,digits)))-sum(digits)
print(subtractProductAndSum(345))
print(subtractProductAndSum(5532))
```

XOR Operation in an Array

You are given an integer n and an integer start.

Define an array nums where nums[i]=start+2*i(0-indexed) and n==nums.length.

Return the bitwise XOR of all elements of nums.

```python
def xorOperation(n:int,start:int)->int:
    result=0
    for i in range(n):
        result^=(start+2*i)
    return result
print(xorOperation(4,3))
print(xorOperation(5,0))

```
